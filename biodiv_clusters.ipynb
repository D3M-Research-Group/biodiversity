```python
import pandas as pd
import numpy as np
# all columns are normalized data
X = pd.read_csv('normalized_final_raw_data.csv')

from k_means_constrained import KMeansConstrained
# make it into an array
X=np.array(X)
# use package
clf = KMeansConstrained(n_clusters = 70, size_min = 7, size_max = 10, random_state=0)
# do the clustering
clf.fit_predict(X)

# array with cluster label for each row
y = clf.labels_

# save to csv
pd.DataFrame(y).to_csv("jag_clusters.csv")
```
